```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
base_url = "https://raw.githubusercontent.com/pranta-iitp/Real-Estate-Property-Price-Prediction-Project/main/houses.csv"
df = pd.read_csv(base_url)
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['link','property_id'],inplace=True)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df.rename(columns={'rate':'pricePerSqft'},inplace=True)
```

```python
df.columns
```

```python
df['society'].value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].isnull().sum()
```

```python
df['society'].unique()
```

```python
# price
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request'].shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df[df['price'] == 'Price on Request'].shape
```

```python
# convert price into crores scales
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
df['price'] = df['price'].str.split(' ').apply(treat_price).astype('float')
```

```python
df.head(5)
```

```python
df['pricePerSqft'] = df['pricePerSqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.info()
```

```python
# bedrooms
df[df['bedRoom'].isnull() == True].shape
```

```python
df = df[df['bedRoom'].isnull() == False]
```

```python
df.shape
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
#bathroom
df[df['bathroom'].isnull() == True]
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
#balcony
df[df['balcony'].isnull() == True]
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
# additionalRoom
df[df['additionalRoom'].isnull() == True].shape
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
# floors
df['noOfFloor'].value_counts()
```

```python
df['noOfFloor'] = df['noOfFloor'].str.split(' ').str.get(0)
```

```python
# changing name because another file has different name
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df.sample(3)
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('not available',inplace=True)
df['facing'] = df['facing'].str.lower()
```

```python
df['area'] = round((df['price']*10000000)/df['pricePerSqft'])
df.insert(loc=1,column='property_type',value='house')
```
